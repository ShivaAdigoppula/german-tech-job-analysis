```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")


```

```python
df = pd.read_csv(r'C:\Users\shiva\Downloads\eda_data.csv\eda_data.csv')
```

```python
# View first 5 rows
df.head()

# Check shape (rows, columns)
df.shape

# Check column names
df.columns

# Dataset info
df.info()

# Check missing values
df.isnull().sum()

# Basic statistics
df.describe()
```

```python
#Top 10 industries by average salary
top_industries = df.groupby('Industry')['avg_salary'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_industries.values, y=top_industries.index, palette='viridis')
plt.xlabel("Average Salary ()")
plt.ylabel("Industry")
plt.title("Top 10 Industries by Average Salary")
plt.show()
```

```python
#Older companies vs startups
plt.figure(figsize=(10,6))
sns.scatterplot(x='age', y='avg_salary', data=df)
plt.xlabel("Company Age (years)")
plt.ylabel("Average Salary ()")
plt.title("Company Age vs Average Salary")
plt.show()
```

```python
#German states with highest salaries
state_salary = df.groupby('job_state')['avg_salary'].mean().sort_values(ascending=False)
plt.figure(figsize=(12,6))
sns.barplot(x=state_salary.index, y=state_salary.values, palette='magma')
plt.xlabel("German State")
plt.ylabel("Average Salary ()")
plt.title("Average Salary by German State")
plt.xticks(rotation=45)
plt.show()
```

```python
#Job description length vs salary
plt.figure(figsize=(10,6))
sns.scatterplot(x='desc_len', y='avg_salary', data=df)
plt.xlabel("Job Description Length (characters)")
plt.ylabel("Average Salary ()")
plt.title("Job Description Length vs Salary")
plt.show()
```

```python
 #Employer-provided vs estimated salaries
plt.figure(figsize=(8,6))
sns.barplot(x='employer_provided', y='avg_salary', data=df, ci=None, palette='coolwarm')
plt.xlabel("Salary Source")
plt.ylabel("Average Salary ()")
plt.title("Employer-provided vs Estimated Salaries")
plt.show()
```

```python
#Number of competitors vs salary
plt.figure(figsize=(10,6))
sns.scatterplot(x='num_comp', y='avg_salary', data=df)
plt.xlabel("Number of Competitors")
plt.ylabel("Average Salary ()")
plt.title("Number of Competitors vs Average Salary")
plt.show()
```

```python
#Salary by job role
plt.figure(figsize=(12,6))
sns.boxplot(x='job_simp', y='avg_salary', data=df, palette='Set2')
plt.xlabel("Job Role")
plt.ylabel("Average Salary ()")
plt.title("Salary Distribution by Job Role")
plt.xticks(rotation=45)
plt.show()
```

```python
#Seniority level vs salary
plt.figure(figsize=(10,6))
sns.boxplot(x='seniority', y='avg_salary', data=df, palette='Set3')
plt.xlabel("Seniority Level")
plt.ylabel("Average Salary ()")
plt.title("Salary by Seniority Level")
plt.show()
```

```python
#Most demanded skills (% of jobs requiring skill)
skill_cols = ['python_yn', 'R_yn', 'spark', 'aws', 'excel']

# Calculate % of jobs requiring each skill
skill_pct = df[skill_cols].mean() * 100  # assuming 1 = skill required, 0 = not required

plt.figure(figsize=(8,6))
sns.barplot(x=skill_pct.index, y=skill_pct.values, palette='pastel')
plt.ylabel("% of Jobs Requiring Skill")
plt.xlabel("Skill")
plt.title("Demand for Skills in Job Market")
plt.show()
```

```python
#Do certain skills lead to higher salaries?
skill_avg_salary = {}
for skill in skill_cols:
    skill_avg_salary[skill] = df.groupby(skill)['avg_salary'].mean().loc[1]  # only jobs requiring the skill

skill_avg_salary = pd.Series(skill_avg_salary).sort_values(ascending=False)

plt.figure(figsize=(8,6))
sns.barplot(x=skill_avg_salary.index, y=skill_avg_salary.values, palette='muted')
plt.ylabel("Average Salary ()")
plt.xlabel("Skill")
plt.title("Average Salary for Jobs Requiring Each Skill")
plt.show()
```

```python
#Company size influence on salary
plt.figure(figsize=(10,6))
sns.boxplot(x='Size', y='avg_salary', data=df, palette='coolwarm')
plt.xlabel("Company Size")
plt.ylabel("Average Salary ()")
plt.title("Salary Distribution by Company Size")
plt.show()
```

```python
#Company rating vs salary
plt.figure(figsize=(10,6))
sns.scatterplot(x='Rating', y='avg_salary', data=df)
sns.regplot(x='Rating', y='avg_salary', data=df, scatter=False, color='red')  # regression line
plt.xlabel("Company Rating")
plt.ylabel("Average Salary ()")
plt.title("Company Rating vs Salary")
plt.show()
```
